```python
people = {
    "first": ["Corey", 'Jane', 'John','Ayoade','Adenike'], 
    "last": ["Schafer", 'Doe', 'Doe',"Olayiwola",'Orimolade'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com', "AyoOlayi@gmail.com","Adeorim@ymail.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
df
```

```python
#to get a series of first name and last name combined

df["first"] + " " + df["last"]
```

```python
#to add the 5 values to a new column
#the apply method can also be used

df["full_name"] = df["first"] + " " + df["last"]
df
```

```python
df
```

```python
#to remove columns we use the drop method on our dataframe
#e.g to drop first and last names from our df

df.drop(columns=["first","last"])
```

```python
#to make the change permanent

df.drop(columns=["first","last"],inplace=True)
```

```python
df
```

```python
#To reverse the process and split the full name into first and last

df["full_name"].str.split()
```

```python
df["full_name"].str.split(" ")
```

```python
#to expand or put into column

df["full_name"].str.split(" ", expand=True)
```

```python
df[["first", "last"]] = df["full_name"].str.split(" ", expand=True)
df
```

```python
#to add or remove rows of data
#first to add a single row of data we use the append method

df.append({"first":"Tony"}, ignore_index=True)
```

```python
df.append({"email":"Tonytte@ymail.com"}, ignore_index=True)
```

```python
#update data
people = {
    "first": ["Caleb", 'Kola'], 
    "last": ["Shun", 'Onit',], 
    "email": ["Calebsh@gmail.com", "KolaOnit@email.com"]
}
df2 = pd.DataFrame(people)
```

```python
df2
```

```python
#to add this to our existing df... observe the index and columns are not the same
#this append method does not require in place

df.append(df2, ignore_index=True)
```

```python
df = df.append(df2, ignore_index=True)
```

```python
#to drop a row
df.drop(index=2, inplace=True)
```
